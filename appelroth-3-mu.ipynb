# Difference schemes for differential equations with a polynomial right-hand side, defining birational correspondences
## Supplement to the article, 25 aug. 2024 

```python
load('fdm.sage')
var("x,z,t")
T=1
u=z-x^2
mu=1
pr=Initial_problem([x,z],[x*z,2*z^2+mu*u],[1,1],T)
pr.latex()
```

Solvable system is
$$
\left \{ \begin{aligned} &
\frac{d}{dt} x = x z , \quad  \frac{d}{dt} z = -x^{2} + 2 \, z^{2} + z , \\ &
x (0)= 1 , \quad z (0)= 1
\end{aligned} \right. 
$$
The solution by Kahan scheme is

```python
sol=cremona_scheme(pr, N=100, field=RR)
```

```python
pl=sol.plot(x,z, points=True)+implicit_plot(x^2==z, (x,-5,5),(z,-5,5), color='red')+implicit_plot(-x^2==z, (x,-5,5),(z,-5,5), color='red')
pl.show(ymax=5, ymin=-5, xmin=-5, xmax=5)
```
