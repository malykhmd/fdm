## Вычисление определителя Кремоны

```python
S=cremona_scheme_general(pqrk,dt,3,ring=QQ[k])
```

## Система Вольтерры-Лотки

```python
var("x,y,t")
pr=Initial_problem([x,y],[x*(2-y),-y*(2-x)],[1,2],500)
pr.latex()
```

$$\left \{ \begin{aligned} &
\frac{d}{dt} x = -x {\left(y - 2\right)} , \quad  \frac{d}{dt} y = {\left(x - 2\right)} y , \\ &
x (0)= 1 , \quad y (0)= 2
\end{aligned} \right. $$

```python
s=cremona_scheme(pr, N=10^3, field=RR)
s.plot(x,y, points=True)
```

```python
u = - (x+y) + 2*log(x*y)
```

```python
s.plot(t,u)
```

```python
s.plot(x,y, points=True)+implicit_plot(u-u.subs([x==1,y==2]),(x,1,4),(y,1,4), color="red") 
```

```python
(s.plot(x,y, points=True)+implicit_plot(u-u.subs([x==1,y==2]),(x,1,4),(y,1,4), color="red")).save("plot-vl-2.pdf")
```

## Случай Ковалевской

```python
A=2
B=2
C=1
x0=1
y0=0
z0=0
M=1
g=10
ics =[0,1,2]+[1,0,0]
ics2 =[0,2,2]+[1,0,0]
T=100
```

```python
var("p,q,r,u,v,w,t")
problem=Initial_problem([p,q,r,u,v,w],[-(C-B)/A*q*r+M*g/A*(y0*w-z0*v), \
                               -(A-C)/B*r*p+M*g/B*(z0*u-x0*w), -(B-A)/C*p*q+M*g/C*(x0*v-y0*u), \
                              r*v-q*w, p*w-r*u, q*u - p*v], ics, T)
problem2=Initial_problem([p,q,r,u,v,w],[-(C-B)/A*q*r+M*g/A*(y0*w-z0*v), \
                               -(A-C)/B*r*p+M*g/B*(z0*u-x0*w), -(B-A)/C*p*q+M*g/C*(x0*v-y0*u), \
                              r*v-q*w, p*w-r*u, q*u - p*v], ics, 1)
problem3=Initial_problem([p,q,r,u,v,w],[-(C-B)/A*q*r+M*g/A*(y0*w-z0*v), \
                               -(A-C)/B*r*p+M*g/B*(z0*u-x0*w), -(B-A)/C*p*q+M*g/C*(x0*v-y0*u), \
                              r*v-q*w, p*w-r*u, q*u - p*v], ics, 5)
I1=A*p^2+B*q^2+C*r^2-2*M*g*(x0*u+y0*v+z0*w)
I2=u^2+v^2+w^2
I3=A*p*u+B*q*v+C*r*w
# Интеграл для случая Ковалевской
I4=(p^2-q^2+(M*g*x0/C)*u)^2 + (2*p*q+(M*g*x0/C)*v)^2
```

```python
problem.latex()
```

$$\left \{ \begin{aligned} &
\frac{d}{dt} p = \frac{1}{2} \, q r , \quad \frac{d}{dt} q = -\frac{1}{2} \, p r - 5 \, w , \quad \frac{d}{dt} r = 10 \, v , \quad \frac{d}{dt} u = r v - q w , \quad \frac{d}{dt} v = -r u + p w , \quad  \frac{d}{dt} w = q u - p v , \\ &
p (0)= 0 , \quad q (0)= 1 , \quad r (0)= 2 , \quad u (0)= 1 , \quad v (0)= 0 , \quad w (0)= 0
\end{aligned} \right. $$

```python
s=cremona_scheme(problem, N=10^3, field=RR)
```
