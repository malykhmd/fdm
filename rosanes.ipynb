## $\wp$-осциллятор

```python
h=d*y^2+a*x^3+b*x^2+c*x
field=FractionField(QQ[a,b,c,d,dt])
f=diff(h,y)
g=-diff(h,x)
```

```python
[A,B]=RosanesMatrices(f,g)
```

```python
latex(A)
```

```python
latex(B)
```

```python
QQ[a,b,c,d,dt][mu](det(A-mu*B)).factor()
```

```python
Mu=field[mu](det(A-mu*B)).roots(multiplicities=False)
Mu
```

```python
[A1,A2,A3]=[ProjectivePoint(((A-mu*B).subs(mu==m)*matrix(p).transpose()).list(), p, field) for m in Mu]
[AA1,AA2,AA3]=[ProjectivePoint((matrix(p)*(A-mu*B).subs(mu==m)).list(), p, field) for m in Mu]
```

```python
[A1,A2,A3]
```

```python
[AA1,AA2,AA3]
```

```python
matrix([list(A1),list(AA3), list(A3)]).det()
```

```python
matrix([list(A2),list(AA2), list(A3)]).det()
```

Точки должны быть базовыми точками модифицированного гамильтониана

```python
def mod_ham(H,x):
    f=[-H.diff(x[1]), H.diff(x[0])]
    mh=H+dt/3*matrix([H.diff(xx) for xx in x])*(1-dt/2*jacobian(f,x))^-1*matrix(2,1,f)
    return SR(mh[0][0]).full_simplify()
```

```python
hm=mod_ham(h,[y,x])
field[x,y](hm.numerator())
```

```python
field[x,y](hm.denominator())
```

```python
Hm=mod_ham(h,[y,x]).subs([x==u/w, y==v/w]).full_simplify()
```

```python
F1=Hm.numerator()
F2=Hm.denominator()
[F1,F2]
```

```python
[F1.subs([i==j for (i,j) in zip([u,v,w],B)]).full_simplify() for B in [A1,A2,A3]+[AA1,AA2,AA3]]
```

```python
[F2.subs([i==j for (i,j) in zip([u,v,w],B)]).full_simplify() for B in [A1,A2,A3]+[AA1,AA2,AA3]]
```
