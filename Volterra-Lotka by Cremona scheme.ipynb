```python
var("x,y,t")
pr=Initial_problem([x,y],[x*(2-y),-y*(2-x)],[1,2],10)
pr.latex()
```

$$
\left \{ \begin{aligned} &
\frac{d}{dt} x = -x {\left(y - 2\right)} , \quad  \frac{d}{dt} y = {\left(x - 2\right)} y , \\ &
x (0)= 1 , \quad y (0)= 2
\end{aligned} \right. 
$$

```python
dt=0.30083
N=400
pr=Initial_problem([x,y],[x*(2-y),-y*(2-x)],[1,2],N*dt)
s=cremona_scheme(pr, N=N, field=RR)
s.plot(x,y, points=True, color='red')
```

```python
s2=erk(pr, N=N, field=RR)
s2.plot(x,y, points=True)
```
